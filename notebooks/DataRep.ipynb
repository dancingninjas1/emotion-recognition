## Understanding the Data Distribution 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### Loading CSV files

```python
df_instrument1 = pd.read_csv('C:/Users/apurv/Desktop/project/Data/acoustic-guitar/annotations_acoustic-guitar.csv')
df_instrument2 = pd.read_csv('C:/Users/apurv/Desktop/project/Data/piano/annotations_piano.csv')
df_instrument3 = pd.read_csv('C:/Users/apurv/Desktop/project/Data/electric-guitar/annotations_electric-guitar.csv')
```

### Creating a combined pandas dataset and couting samples for each emotion per instrument

```python
# Combine data from all instruments
df_combined = pd.concat([df_instrument1, df_instrument2, df_instrument3])

# Combine "steelstring-guitar" and "classical-guitar" under a single category "guitar"
df_combined['instrument'] = df_combined['instrument'].replace({'steelstring-guitar': 'guitar', 'classical-guitar': 'guitar'})

# Create a table of the data showing count of each emotion per instrument
table = pd.crosstab(df_combined['emotion'], df_combined['instrument'])

# Display the table
print(table)
```

### Visualizing the Data

```python


# Count the occurrences of each emotion
emotion_counts = df_combined['emotion'].value_counts()

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

# Use barplot to show count of each emotion for combined data
sns.barplot(data=df_combined, x=emotion_counts.index, y=emotion_counts.values, ci=None)

plt.title('Count of Samples for Each Emotion')
plt.xlabel('Emotion')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Combine data from all instruments

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

sns.barplot(data=df_combined, x='emotion', y='emotional_intensity', hue='instrument', ci=None)

plt.title('Emotional Intensity Grouped Bar Graph')
plt.xlabel('Emotion')
plt.ylabel('Emotional Intensity')
plt.legend(title='Instrument')
plt.show()
```
